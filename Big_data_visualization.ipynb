Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps with the highest number of installs.

```python
df['Installs'] = pd.to_numeric(df['Installs'].str.replace('[+,]', '', regex=True), errors='coerce')
install_frequency = df.groupby('App')['Installs'].sum().reset_index()
install_frequency_sorted = install_frequency.sort_values(by='Installs', ascending=False)
top_10_apps = install_frequency_sorted.head(10)
print(top_10_apps)
```

Q15. A company wants to launch a new app on the Google Playstore and has asked you  o analyze the Google Playstore dataset to identify the most popular app categories. How would you approach this task, and what features would you analyze to make recommendations to the company?

```python
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')
category_install_frequency = df.groupby('Category')['Installs'].sum().reset_index()
category_install_frequency_sorted = category_install_frequency.sort_values(by='Installs', ascending=False)
most_popular_categories = category_install_frequency_sorted
print(most_popular_categories)
```

Q16. A mobile app development company wants to analyze the Google Playstore dataset to identify the most successful app developers. What features would you analyze to make recommendations to the company, and what data visualizations would you use to present your findings?

```python
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')
developer_install_frequency = df.groupby('Genres')['Installs'].sum().reset_index()
most_successful_developers = developer_install_frequency.sort_values(by='Installs', ascending=False)
print(most_successful_developers.head(10))
```
